```
from sympy import*
```

```
x = Symbol('x, y')
f1 = x ** 3 - 50 * x
f2 = -x ** 4 + 88 * x ** 2 - 241
```

```
f1
```

```
f2
```

```
# Точки пересечения этих функций

i = f1 - f2
```

```
i
```

```
x1, x2, x3, x4 = solve(i)
```

```
x1=x1.evalf(chop=True)
x2=x2.evalf(chop=True)
x3=x3.evalf(chop=True)
x4=x4.evalf(chop=True)
x1 = round(x1, 3)
x2 = round(x2, 3)
x3 = round(x3, 3)
x4 = round(x4, 3)
```

```
x1, x2, x3, x4
```

```
y1=f1.subs(x,x1)
y2=f1.subs(x,x2)
y3=f1.subs(x,x3)
y4=f1.subs(x,x4)
```

```
y1, y2, y3, y4
```

```
# графики функций в одной системе координат
i = plot((f1), (f2), show=False)
i[0].line_color = 'blue'
i[1].line_color = 'green'
i.show()
```

```
# график общей функции
j = plot((f1+f2), show=False)
j[0].line_color = 'red'
j.show()
```
